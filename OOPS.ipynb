```python
# We learnt about objects, now we will learn about the 4 major concepts in Object based or Object oriented programming. this is also called OOPS

# 1. Inheritance
# 2. Abstraction
# 3. Encapsulation
# 4. Polymorphism


# Do we know other shapes apart from rectangles?

# Square
# any others?
# Circle

#What are all of them? They are shapes.

class Shape:
    def __init__(self,colour):
        self.colour = colour

    def area(self):
        print('This is printing from inside Shape')

    def print_area(self):
        print(self.area())
        print(self.colour)


class Rectangle(Shape):
    def __init__(self,colour, length, breadth):
        super().__init__(colour)
        self.length = length
        self.breadth = breadth

    #overriding method in base class. This is called polymorphism. Depending on the object you create you get the method
    def area(self):
        return self.length * self.breadth  

class Circle(Shape):
    pi = 3.14
    def __init__(self,colour, radius):
        super().__init__(colour)
        self.radius = radius        

    def area(self):
        return self.pi * (self.radius ** 2)  


a = Rectangle('blue', 25, 10)
b = Shape('black')
c = Circle('yellow', 10)

#a.print_area()
#b.print_area()
c.print_area()
```

```python
# Abstraction provides a programmer a way to hide all the irrelevant
# data/process of an application in order to reduce complexity
# and increase the efficiency of the program.

# Abstract class is a blueprint or contract for other classes
# Every method in this class will be empty, so you will be forced to implement something in your derived class
# To create an abstract class you have to import the module ABC. And its annotators called @abstractmethod

from abc import ABC, abstractmethod


class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def print_area(self):
        pass

    @abstractmethod
    def no_of_sides(self):
        pass


class Rectangle(Shape):
    def __init__(self, breadth, length):
        self.breadth = breadth
        self.length = length

    # overriding abstract method
    def area(self):
        return self.breadth * self.length

    # overriding abstract method
    def no_of_sides(self):
        return 4

    # overriding abstract method
    def print_area(self):
        print(self.area())
        print(self.no_of_sides())


a = Rectangle(10, 20)

a.print_area()
```

```python
# Special methods/Dunder methods

from abc import ABC, abstractmethod


class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def print_area(self):
        pass

    @abstractmethod
    def no_of_sides(self):
        pass


class Rectangle(Shape):
    def __init__(self, breadth, length):
        self.breadth = breadth
        self.length = length

    # overriding abstract method
    def area(self):
        return self.breadth * self.length

    # overriding abstract method
    def no_of_sides(self):
        return 4

    # overriding abstract method
    def print_area(self):
        print(self.area())
        print(self.no_of_sides())

    def __str__(self):
        return f'This is a rectangle object. It has {self.no_of_sides()} sides. Its area is {self.area()}'

    def __len__(self):
        return self.no_of_sides()

    def __del__(self):
       print("Rectangle has been deleted")

a = Rectangle(10, 20)

print(a)
print(len(a))
a.print_area()

del a

try:
    print(a)
except NameError:
    "The object has been deleted"
finally:
    "You can try printing something else in this area"
    print('Bye for now')

```
